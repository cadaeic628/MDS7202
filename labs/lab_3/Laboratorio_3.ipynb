**Respuesta:**

````python
def to_snake_case(s: str) -> str:
    s = s.strip().lower()
    s = s.replace("/", " ").replace("-", " ").replace("'", "")
    s = re.sub(r"[^a-z0-9]+", "_", s)
    s = re.sub(r"_+", "_", s).strip("_")
    return s
````

````python
def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    df_clean = df.copy()

    # Estandarizar nombres de columnas
    df_clean.columns = [to_snake_case(col) for col in df_clean.columns]

    # Renombrar race_ethnicity -> ethnicity
    if "race_ethnicity" in df_clean.columns:
        df_clean = df_clean.rename(columns={"race_ethnicity": "ethnicity"})

    # Eliminar duplicados exactos
    df_clean = df_clean.drop_duplicates()

    # Convertir scores a numéricos
    score_cols = ["math_score", "reading_score", "writing_score"]
    for col in score_cols:
        df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")

    # Eliminar filas con nulos
    df_clean = df_clean.dropna()

    # Filtrar notas válidas entre 0 y 100
    for col in score_cols:
        df_clean = df_clean[(df_clean[col] >= 0) & (df_clean[col] <= 100)]

    return df_clean.reset_index(drop=True)
````

````python
df_cleaned = clean_data(df_grades)
df_cleaned
````
