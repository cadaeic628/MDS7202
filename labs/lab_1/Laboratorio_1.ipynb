### Extra: ¿Qué es el CI y qué verifican estas herramientas?

**CI (Integración Continua)** es una práctica en la que cada vez que se sube código al repositorio (mediante un push o un Pull Request), se ejecutan automáticamente una serie de verificaciones para asegurar que el código cumple ciertos estándares de calidad. En este curso, el CI está configurado en `.github/workflows/ci.yml` y es ejecutado por GitHub Actions.

Las herramientas que configuramos cumplen roles complementarios:

| Herramienta            | ¿Qué verifica?                                                                                                                   |
| ---------------------- | -------------------------------------------------------------------------------------------------------------------------------- |
| **`ruff` (linter)**    | Detecta errores de estilo, imports no usados, variables indefinidas, patrones propensos a bugs y sintaxis desactualizada         |
| **`ruff` (formatter)** | Formatea el código automáticamente para que tenga un estilo consistente (espacios, saltos de línea, comillas, etc.)              |
| **`pre-commit`**       | Ejecuta un conjunto de revisiones automáticamente _antes de cada commit_, impidiendo subir código que no pase las verificaciones |

En la práctica: cuando hagan `git commit`, `pre-commit` correrá `ruff` sobre sus archivos. Si hay problemas, el commit será bloqueado y deberán corregirlos antes de continuar. Esto asegura que todo el código que llegue al repositorio sea consistente y libre de errores básicos.

#### Pregunta 1 (0.5 pts)

Dada una lista de $n$ elementos, no necesariamente ordenada, defina funciones que busquen:

- El elemento menor.
- El elemento mayor.
- El promedio de los elementos en la lista.

Ejemplo:

```python
lista_n = [12,4,34,1,2,52,1,54,98,5,235,76,57,87,123,32,32]
menor = 1
mayor = 235
```


````python
def obtener_estadisticas(lista_n):
    # Verificamos que la lista no esté vacía para evitar errores
    if not lista_n:
        return None, None, None

    menor = lista_n[0]
    mayor = lista_n[0]
    suma = 0
    cantidad = 0

    for numero in lista_n:
        # Encontrar el menor
        if numero < menor:
            menor = numero

        # Encontrar el mayor
        if numero > mayor:
            mayor = numero

        # Acumular para el promedio
        suma += numero
        cantidad += 1

    promedio = suma / cantidad

    return menor, mayor, promedio
````

#### Pregunta 2 (0.5 pts)

Genere una función que obtenga la moda de una lista usando diccionarios.


````python
def obtener_moda(lista):
    # Crear diccionario para contar frecuencias
    frecuencias = {}

    for elemento in lista:
        if elemento in frecuencias:
            frecuencias[elemento] += 1
        else:
            frecuencias[elemento] = 1

    # Encontrar la frecuencia máxima manualmente
    max_frecuencia = None
    moda = None

    for elemento in frecuencias:
        if max_frecuencia is None:
            max_frecuencia = frecuencias[elemento]
            moda = elemento
        else:
            if frecuencias[elemento] > max_frecuencia:
                max_frecuencia = frecuencias[elemento]
                moda = elemento

    return moda
````

#### Pregunta 3 (0.5 pts)

Vea si un número es palíndromo o no. Los números palíndromos son aquellos que al leerlos al revés y al derecho son iguales, por ejemplo el 123321 o el 98789.


````python
def es_palindromo(numero):
    original = numero
    reverso = 0

    while numero > 0:
        digito = numero % 10  # ultimo digito
        reverso = (reverso * 10) + digito  # Desplazamos el número invertido a la izquierda y sumamos el nuevo dígito
        numero = numero // 10  # Le quitamos el último dígito al número original

    return original == reverso
````

#### Pregunta 4 (0.5 pts)

Dado un texto, conviértalo en estilo Camel Case, vea el siguiente ejemplo:

    Texto inicial: “Este es un texto de ejemplo para mostrar el Camel Case”.

    Texto final: “EsteEsUnTextoDeEjemploParaMostrarElCamelCase”.


````python
def convertir_camel_case(texto):
    # Creamos los diccionarios de mapeo para mayúsculas y minúsculas
    letras_min = "abcdefghijklmnopqrstuvwxyz"
    letras_may = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

    # map_mayusculas: {'a': 'A', 'b': 'B', ...}
    map_mayusculas = dict(zip(letras_min, letras_may, strict=True))
    # map_minusculas: {'A': 'a', 'B': 'b', ...}
    map_minusculas = dict(zip(letras_may, letras_min, strict=True))

    resultado = ""
    nueva_palabra = True

    for caracter in texto:
        if caracter == " ":
            nueva_palabra = True
        else:
            if nueva_palabra:
                # Busca el caracter en el diccionario para pasarlo a mayúscula.
                # Si no está devuelve el mismo caracter.
                resultado += map_mayusculas.get(caracter, caracter)
                nueva_palabra = False
            else:
                # Busca el caracter para pasarlo a minúscula.
                resultado += map_minusculas.get(caracter, caracter)

    return resultado
````

#### Pregunta 1 (0.5 pts):

Un monto inicial de dinero (por ejemplo un millón de pesos) se puede depositar en un banco con un interés anual de 5%. Esto significa que cumplido un año se dispondrá de \\$1.050.000. Si se deposita por dos años, entonces al finalizar el 2º año se dispondrá de \\$1.102.500, es decir, \\$1.050.000 reajustado en 5%.

Al respecto, si se invoca a la función `calc_capital_final(1000000, 5, 2)` entrega el resultado `1050000.0`.

Escriba la función `calc_capital_final`

$$C_f = C_i(1+i)^n$$

- $C_f$ = Capital Final
- $C_i$ = Capital Inicial
- $i$ = Tasa de interés
- $n$ = Periodo del ahorro

Además, incluya tests que certifiquen el correcto funcionamiento de la función.


````python
def calc_capital_final(capital_inicial, tasa_interes, años):
    tasa = tasa_interes / 100
    return round(capital_inicial * (1 + tasa) ** años, 1)


# TESTS

# Caso base 1 año
assert calc_capital_final(1000000, 5, 1) == 1050000.0

# 2 años (interés compuesto)
assert calc_capital_final(1000000, 5, 2) == 1102500.0

# Sin interés
assert calc_capital_final(1000000, 0, 3) == 1000000.0

# Sin años
assert calc_capital_final(1000000, 5, 0) == 1000000.0

# Caso general
assert calc_capital_final(500000, 10, 3) == 665500.0
````

#### Pregunta 2 (0.5 pts):

Un anagrama es una palabra que resulta de la transposición de letras de otra palabra o frase.

Por ejemplo, las palabras “caso” y “caos” sonan agramas de “saco” y las palabras “acaso” y “ocaso” no son anagramas de “caso”. Las anagramas poseen las mismas letras, con la misma cantidad de apariciones, pero en un orden diferente

Escriba la función `def anagrama(x,y)` que reciba dos palabras y entregue `True` si son anagramas y `False` si no.


````python
def anagrama(x, y):
    # Pasamos todo a minúsculas por si acaso
    x_limpio = x.lower()
    y_limpio = y.lower()

    # Si no tienen la misma cantidad de letras, es imposible que sean anagramas
    if len(x_limpio) != len(y_limpio):
        return False

    # Diccionarios para guardar el conteo de cada letra
    conteo_x = {}
    conteo_y = {}

    # Contamos las letras de la primera palabra
    for letra in x_limpio:
        conteo_x[letra] = conteo_x.get(letra, 0) + 1

    # Contamos las letras de la segunda palabra
    for letra in y_limpio:
        conteo_y[letra] = conteo_y.get(letra, 0) + 1

    # Si los diccionarios son exactamente iguales, devolvemos True
    return conteo_x == conteo_y


# Ejemplos de uso:
print(anagrama("caso", "caos"))  # Debería imprimir True
print(anagrama("saco", "caso"))  # Debería imprimir True
print(anagrama("acaso", "ocaso"))  # Debería imprimir False
````

### Crear el Pull Request

1. Subir su código ejecutando `git push origin lab_1`
2. Ir a su repositorio de github y ir a la sección **Pull requests -> New pull request**
3. Elegir de qué rama a qué rama es el PR. La rama base debería ser **main** y la rama compare debería ser **lab_1**
4. Crear el Pull Request con el siguiente formato de título y descripción:

**Título:** `[Lab 1] Entrega Laboratorio 1`

**Descripción:**

```
# Laboratorio 1: Python y Git

Entrega del primer laboratorio del curso MDS7202.
Se incluyen ejercicios de Python (listas, strings, funciones) y ejercicios de Git (ramas, merge, blame).

**Fecha:** DD/MM/AAAA HH:MM
**Integrantes:**
- Nombre Apellido 1
- Nombre Apellido 2
```

5. Verificar que el pipeline de CI (`.github/workflows/ci.yml`) pasa correctamente antes de aprobar el PR.
6. Aprobar el pull request y copiar el vínculo de este. Si perdieron la ventana, pueden ir a la sección **Pull requests** y luego a _closed_ para encontrarlo.
7. Subir el vínculo del PR a la tarea de u-cursos.
8. En local, cambiarse a la rama main (`git checkout main`) y luego `git pull`. En esta primera versión, su rama main y lab_1 son iguales por lo que no hará nada. Pero en laboratorios posteriores será necesario para evitar conflictos.

Deberán realizar este mismo procedimiento para cada laboratorio. Recomendamos primero descargar el notebook en su repo local en su respectiva carpeta, crear la rama del laboratorio y trabajar directamente en esa rama, realizando commits en cada paso que den, lo cual les facilitaría su trabajo en caso de que se equivoquen.

**No olvidar:** antes de crear una rama nueva siempre hacerlo desde main, y siempre hacer `git pull` antes de crear la nueva rama.
